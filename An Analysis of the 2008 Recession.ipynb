```python
import datetime
from datlib.FRED import *
from datlib.plots import *
data_codes  = {"Inflation Rate":"T10YIE",
               "Federal Spending":"FGEXPND",
              "Market Value of Real Estate":"USSTHPI",
              "Consumer Price Index":"CPILFESL",
              "Real Gross Domestic Product":"GDPC1",
              "Unemployment Rate":"UNRATE",
              "Mortgage Delinquency Rate":"DRSFRMACBS",
              "Mortgage-Backed Securities held by FED":"WSHOMCB",
              "Monthly Supply of Houses in the United States":"MSACSR",
              "Fannie Mae Debt":"MDOTHFRAFNMA"}
start = datetime.datetime(1975, 1, 1)
end = datetime.datetime.today()
```

```python
data = gather_data(data_codes = data_codes, 
                   start = start, 
                   end = end, 
                   freq = "Q")
```

# Introduction

The 2008 economic recession, commonly referred to as The Great Recession, is the economic downturn that was experienced in late 2007 to early 2009. This downturn can be seen in the chart below with the blue line depicting the decrease in Real GDP and the orange line depicting the sharp increase in unemployment around the year of 2008. Washington Post highlights this issue by stating, "The crisis was the worst U.S. economic disaster since the Great Depression. In the United States, the stock market plummeted, wiping out nearly 8 trillion dollars in value between late 2007 and 2009. Unemployment climbed, peaking at 10 percent in October 2009. Americans lost 9.8 trillion dollars in wealth as their home values plummeted and their retirement accounts vaporized." In this analysis, I will lay out varying perspectives that explain the causes of this recession, and I will follow it up with an account of my own sentiments regarding the potential causation.

```python
data_to_plot = data[["Real Gross Domestic Product", "Unemployment Rate"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20), secondary_y = "Unemployment Rate")
plt.rcParams.update({"font.size":32})
```

# Varying Perspectives

Fox Business stated, "The worst financial crisis since the Great Depression was triggered by overheating in the housing markets. Banks and other lenders approved mortgages, sometimes to borrowers with poor credit histories, driving up home prices to astronomical levels. Banks then sold the risky mortgage-backed securities to other financial institutions." To echo Fox Business, there seems to be a general consensus that the 2008 recession was caused by the subprime mortgage crisis. Essentially, this is the term used to explain the actions that banks took in the early 2000's when they provided home loans to buyers who were not qualified to purchase those homes. These mortgages were packaged together and sold as securities. These securities were then graded to explain the risk, but this process was somewhat corrupted at the time. Essentially, what ended up happening is that many home-owners defaulted on their loans. This can be seen by the following data that depicts the delinquency rate on single-family residential mortgages.

```python
data_to_plot = data[["Mortgage Delinquency Rate"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20))
plt.rcParams.update({"font.size":32})
```

In turn, this high delinquency rate lead to many individuals foreclosing on their homes. This drastically increased the housing supply in the US:

```python
data_to_plot = data[["Monthly Supply of Houses in the United States"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20))
plt.rcParams.update({"font.size":32})
```

Although there is this general consensus that the cause was due to the mortgage crisis, there are differing perspectives as to what caused the subprime mortgage crisis. One of these perspectives is that Wall Street and the big financial firms were irresponsible and are at fault. Sun Sentinal weighed in on the behavior of Wall Street. They stated, "The great irony of the 2008 financial collapse is that Wall Street, whose reckless risk-taking drove the financial system over the precipice, suffered very few, if any, consequences for its actions. The crisis cost millions of people their jobs and their homes, devastated cities and towns across the nation and stripped away trillions of dollars in household wealth from the middle class. But the big banks barely skipped a beat, paying no real economic, legal or political price for their misconduct." Another outlet that blamed Wall Street for the 2008 recession was NPR. They explained, "Morgenson focuses on the managers of Fannie Mae, the government-supported mortgage giant. She writes that CEO James Johnson built Fannie Mae "into the largest and most powerful financial institution in the world." But in the process, Morgenson says, the company fudged accounting rules, generated big salaries and bonuses for its executives, used lobby and campaign contributions to bully regulators, and encouraged the risky financial practices that led to the crisis."

Another theory for the causation of this economic collapse is the idea that there wasn't enough government regulation surrounding financial firms. According to the Center for American Progress, "Government policies that make it more affordable to buy a home were not responsible for the crisis. In fact, consumers who already had mortgages and who had built up equity in their homes were more likely to be targeted for predatory subprime loans than first-time homebuyers. Instead of too much government, it was the lack of sufficient government oversight in key areasincluding consumer protection, private label mortgage securitization, bank capitalization, and financial marketsthat transformed a housing bubble into a global financial crisis." The Balance also cited the lack of government regulation as the main cause of this recession. According to The Balance, "Deregulation in the financial industry was the primary cause of the 2008 financial crash. It allowed speculation on derivatives backed by cheap, wantonly-issued mortgages, available to even those with questionable creditworthiness."

There is also a belief that the government incentivized lenders to give out sub-par loans. Anthony Randozzo and Carson Young with the Reason Foundation argue that Fannie Mae and Freddie Mac had a large impact on the financial crisis. They wrote, "In the late 1990s, Fannie and Freddie began to rapidly increase the purchasing of mortgages and securitized mortgages (also known a mortgage-backed securities) from the private sector. The banks recognized that the increased demand was because of the government directives and it meant they could issue more mortgages with lowered lending standards, collect the fees associated with origination, and then sell the loans to the GSEs without holding on to the risk." The Cato institute also referenced this issue. According to the Cato Institute, "in 1992 Congress started passing affordable housing goals and over the years continued to increase those goals. The banks were told to make an increasing percentage of their loans to buyers who would not normally qualify, and Fannie and Freddie were in turn pressured or required to buy more and more of these nontraditional or subprime mortgages. Down payment requirements became easier and easier, and finally, in many cases, zero. Credit and employment history requirements became less and less strict, finally leading to the nodoc loan."

There were also some perspectives that seemed to go against popular beliefs. The Heritage Foundation offered one of these perspectives. According to the Heritage Foundation, "The key policy failure was most likely regulators' decision in favor of bailing out Bear Sterns, a (smaller) competing investment bank, rather than their decision against saving Lehman. The Bear Sterns bailout set the expectation that Lehman would also be bailed out. Going forward, investors and creditors surely expected, at the very least, the government to minimize their losses."

Lastly, Pro Market analyzed who the American population at large felt was at fault. Pro Market explained, "The financial crisis has shaken the economists view of the rationality of individuals and efficiency of markets. After regulation, the most highly rated causes of the crisis were irrational beliefs (on house prices or risk) and corrupt incentives (fraud in mortgages and credit rating agencies)."

# My Perspective

Although I think that a mixture of causes led to the Great Recession, I found the arguments about affordable housing goals set by congress to be the prime driver of the housing bubble that led to millions losing their jobs. Ultimately, this policy misaligned incentives in favor of banks and large financial firms giving out subprime loans, that were often insured by the government, in order to yield higher interest rates, and, ultimately, make more money. Also, firms, such as Fannie Mae, purchased massive amounts of mortgage backed securities due to this government influence. Since Fannie Mae was tremendously overleveraged, they found themselves in a lot of debt. This can be seen by the following graph highlighting their debt:

```python
data_to_plot = data[["Fannie Mae Debt"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20))
plt.rcParams.update({"font.size":32})
```

In addition, I agree with the idea that the recklessness of Wall Street was one of the main contributors to this problem; however, I think the expectation these firms had of government support is what led them to justify taking senseless risk.

Another discussed cause of this recession can be put on the shoulders of the American population, in general. Many individuals wanted to make money off of the quickly rising prices of housing. When the housing market crashed in 2009, many Americans were left with mortgages that they couldn't afford. This change in housing prices in relation to the Consumer Price Index can be seen below:

```python
data_to_plot = data[["Market Value of Real Estate", "Consumer Price Index"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20))
plt.rcParams.update({"font.size":32})
```

Although I believe that a problem was posed with an expectance of government bailouts for financial firms, I disagree with the beliefs that were shared in the Heritage article. Essentially, I don't believe that the bailout of Bear Stearns was the main policy failure in this scenario. This analysis seems to miss the point that Bear Stearns was going bankrupt due to the other factors that caused the subprime mortgage crisis.

Lastly, while I focused mostly on the causes of this recession, there were also many lasting implications. For example, before the Great Recession, the Federal Reserve held no mortgage-backed securities. This quickly changed when the Fed used open market operations in 2009 to begin purchasing these securities.

```python
data_to_plot = data[["Mortgage-Backed Securities held by FED"]]
plot_lines(df = data_to_plot, linewidth = 5, figsize = (40, 20))
plt.rcParams.update({"font.size":32})
```

The many different perspectives suggest that the causes could vary from regulatory issues to problems with consumer and corporate risk taking. Ultimately, most people can agree that this economic downturn resulted from poor financial decisions and even worse policy decisions.
